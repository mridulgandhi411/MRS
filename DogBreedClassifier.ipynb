```python
height=100
width=100
channels=3
X_valid=None
X_valid=np.ndarray(shape=(len(files_to_valid),height,width,channels),dtype=np.float32)
print("Shape of validation data:",X_valid.shape)

#for i in train_images:
 #   j='D:/dogImages/train/'+i
  ##     img=load_img(j+'/'+k)
a=range(0,len(valid_images))
count=0
var=0
for i in names2:
    for k in i:
        img=load_img('D:/dogImages/valid/'+valid_images[a[count]]+'/'+k)
        img=img.resize((width,height),Image.ANTIALIAS)
        x=img_to_array(img)
        X_valid[var]=x
        var+=1
    count+=1
    
  
```

```python
target_series=pd.Series(y_train)
one_hot=pd.get_dummies(target_series,sparse=True)
y_train=np.asarray(one_hot)
```

```python
target_series=pd.Series(y_test)
one_hot=pd.get_dummies(target_series,sparse=True)
y_test=np.asarray(one_hot)
```

```python
target_series=pd.Series(y_valid)
one_hot=pd.get_dummies(target_series,sparse=True)
y_valid=np.asarray(one_hot)
```

```python
from keras.models import Sequential 
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D
from keras.layers import GlobalAveragePooling2D

model=Sequential()
model.add(Conv2D(filters=32,kernel_size=5,padding='same',activation='relu',input_shape=(100,100,3)))
model.add(MaxPooling2D(pool_size=2))

model.add(Conv2D(filters=64,kernel_size=5,activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.3))

model.add(Conv2D(filters=32,kernel_size=5,padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=2))

model.add(Conv2D(filters=64,kernel_size=5,activation='relu'))
model.add(MaxPooling2D(pool_size=2))


model.add(Conv2D(filters=32,kernel_size=5,padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.2))


model.add(GlobalAveragePooling2D())

model.add(Dense(1000,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(133,activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint,EarlyStopping

checkpointer=ModelCheckpoint(filepath='model.dogbreed.hdf5',verbose=1,save_best_only=True)
early_stop=EarlyStopping(monitor='val_loss',patience=4,verbose=1)
cnn=model.fit(X_train,y_train,batch_size=100,epochs=20,validation_data=(X_valid,y_valid),callbacks=[checkpointer,early_stop],verbose=1,shuffle=True)
```

```python
model.load_weights('model.dogbreed.hdf5')
```

```python
score=model.evaluate(X_test,y_test,verbose=0)
print("Test Accuracy: ",score[1])
```
