```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('./datasets/scrobbler-small-sample.csv')
print(data.head())
data['artist_offset'].unique()
```

```python
artists=pd.read_csv('./datasets/artists.csv',header=None,names=['artist_names'])
artists.head()
```

```python
from scipy.sparse import coo_matrix

rows=data['user_offset']
cols=data['artist_offset']
count=data['playcount']

matrix=coo_matrix((count,(rows,cols)))
sparse_matrix=pd.DataFrame(matrix.toarray(),columns=artists['artist_names'])
sparse_matrix
```

```python
df=sparse_matrix.T
df
```

```python
from sklearn.decomposition import NMF
```

```python
model=NMF(n_components=7)
model.fit(df)
nmf_features=model.transform(df)
nmf_features.shape
```

```python
from sklearn.preprocessing import normalize
nom_features=normalize(nmf_features)
df1=pd.DataFrame(nom_features,index=artists['artist_names'])
artist=df1.loc['Eminem']
similarities=df1.dot(artist)
print(similarities.nlargest())
```

```python
artist=df1.loc['The Killers']
similarities=df1.dot(artist)
print(similarities.nlargest())
```

## We can also sparse the matrix by pivot_table()

```python
data=pd.read_csv('./datasets/scrobbler-small-sample.csv')
data.head()
```

```python
data1=data.pivot_table('playcount',index='artist_offset',columns='user_offset')
```

```python
data1.fillna(0,inplace=True)
data1
```

```python
model=NMF(n_components=7)
model.fit(data1)
nmf_features=model.transform(data1)
nmf_features.shape
```

```python
x=artists['artist_names']
```

```python
nom_features=normalize(nmf_features)
df=pd.DataFrame(nom_features,index=x)
artist=df.loc['Simon & Garfunkel']
similiarities=df.dot(artist)
print(similarities.nlargest())
```
